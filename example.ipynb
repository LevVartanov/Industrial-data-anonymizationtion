```python
import library
```

```python
import pandas as pd
```

Downloading dataset

```python
df = pd.read_csv(('adult.csv'))
```

```python
df.replace({'<=50K': 0, '>50K': 1}, inplace=True)
```

Separating target feature

```python
X = df.drop("income", axis = 1)
y = df["income"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state = 44)
```

Initializing (presetted) model

```python
model = library.EconomicTabularAnonymizer()
```

Encoding (with presetted parrameters for this dataset)

```python
X = model.encode(X)
```

```python
model.fit(X, y)
```

Generating anonymized data

```python
X_gen, y_gen = model.generate()
```
